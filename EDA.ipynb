```python
import pandas as pd

# Load the census and GDP data
census_df = pd.read_csv('india-districts-census-2011.csv')
gdp_df = pd.read_csv('gdp_AndhraPradesh1.csv')

census_df.head()
census_df.info()
census_df.isnull().sum()
```

```python
for col in census_df.columns:
    print(f"{col}: {census_df[col].nunique()} unique values")

census_df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Only include columns that exist
income_cols = ['Power_Parity_Above_Rs_545000']  # Adjusted to your dataset

for col in income_cols:
    plt.figure(figsize=(12, 6))
    sns.histplot(census_df[col], bins=30, kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel('Number of Households')
    plt.ylabel('District Count')
    plt.show()
```

```python
# Correct column name: 'State Name'
state_income = census_df.groupby('State Name')[income_cols].sum().sort_values(by='Power_Parity_Above_Rs_545000', ascending=False)

# Visualization
state_income['Power_Parity_Above_Rs_545000'].plot(kind='bar', figsize=(12, 6), color='lightgreen')
plt.title('High Income Population (Rs. 5.45L+) by State')
plt.xlabel('State')
plt.ylabel('Number of Households')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```
